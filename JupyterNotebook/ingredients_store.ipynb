```python
import pickle
import pandas as pd
import nltk
#make sure these library items are downloaded
#nltk.download('punkt')
#nltk.download('averaged_perceptron_tagger')
#nltk.download('wordnet')
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.corpus import wordnet as wn


# read python dict back from the file
pkl_file = open('countedingredientwords.pkl', 'rb')
countedingredients = pickle.load(pkl_file)
pkl_file.close()

#print(countedingredients)

#masteringredient=[]
labelledwords={}
rarewords =[]
#initialize with some manual terms
mystopwords=["Oven","it","Cuisine"]
regularwords = []

for key,value in countedingredients.items():
    # assumes words that appear over a certain threshold to be insignificant, here 73 times
    if value >=73:
        mystopwords.append(str(key))
        labelledwords[key.lower()]=(value,"common")
    # marking all unique words that appear only once to be significant/rare
    if value<2:
        rarewords.append(str(key))
        labelledwords[key.lower()]=(value,"rare")
    elif value >=2 and value <73:
        regularwords.append(str(key))
        labelledwords[key.lower()]=(value,"regular")

with open('./Dataset/unique_cuisines.txt', 'r') as file:
     cuisinescontent = file.readlines()
file.closed

#manually adding country related words in case these are significant
cuisines = []
for cuisine in cuisinescontent:
    cuisines.append(str(cuisine).lower())

print("I am processing the rare and medium words to see whether they are verbs, numbers or adjectives")
print("here are the cuisines") 
for word in cuisines:
    word = str(word)
    print(word)

#only add adjectives to stop words if they are not a cuisine word 

for word in rarewords:
    processed = pos_tag([word])
#     print(word, processed)
    #split out the tag identifying the word type
    tag = ", ".join(t for w, t in processed)
 #   print(tag)
    if str(tag)=="CD" or str(tag)=="VBN" or str(tag)=="JJ":
        word = word.lower()  
        if word in cuisines:
            print("This word is a cuisine, I will not add it to stop words!")
            print(word)
            try:
                labelledwords[word]=(labelledwords[word],"cuisine")
            except:
                labelledwords[word.lower()]=(labelledwords[word],"cuisine")

        elif word.lower() not in cuisines:
            print("The word is not in the list!")
            #print(word)
            #print("I found a number or verb or adjective which is not a cuisine")
            mystopwords.append(word)
            try:
                labelledwords[word]=(labelledwords[word],"tagged")
            except:
                labelledwords[word.lower()]=(labelledwords[word],"tagged")
                
for word in regularwords:
    processed = pos_tag([word])
#     print(word, processed)
    #split out the tag identifying the word type
    tag = ", ".join(t for w, t in processed)
 #   print(tag)
    if str(tag)=="CD" or str(tag)=="VBN" or str(tag)=="JJ":
        word = word.lower()  
        if word in cuisines:
            print("This word is a cuisine, I will not add it to stop words!")
            print(word)
            try:
                labelledwords[word]=(labelledwords[word],"cuisine")
            except:
                labelledwords[word.lower()]=(labelledwords[word],"cuisine")

        else:
            print("The word is not a cuisine!")
            print(word)
            #print("I found a number or verb or adjective which is not a cuisine")
            try:
                labelledwords[word]=(labelledwords[word],"tagged")
                mystopwords.append(word)
            except:
                labelledwords[word.lower()]=(labelledwords[word],"tagged")
                mystopwords.append(word)


print("here are the cuisines") 
for word in cuisines:
    word = str(word)
    print(word)
    if word in mystopwords:
        print("removing word")
        print(word)
        mystopwords.remove(word)

                
print("labelledwords")
print(labelledwords)
print("mystopwords") 
print(mystopwords) 

print("Here is my dictionary labelling how important each word is")
#print(labelledwords)

#do not run this unless needed! writing to countedingredientwords
#output = open('labelledwords.pkl', 'wb')
#pickle.dump(labelledwords, output)
#output.close()
'''
ingredientword=[]
occurrences=[]
label=[]

for key, value in labelledwords.items():
    ingredientword.append(str(key))
    occurrences.append(value[0])
    label.append(value[1])
    
df = pd.DataFrame({'ingredientword': ingredientword, 'Occurrences':occurrences,'label':label}).sort_values(by=['label'],ascending=False)
df
'''
```
