```python
# work with ingredients to identify most common words /good candidates for stemming and classifying
import re
# Load in the pandas library
import pandas as pd
import numpy as np


with open('./Dataset/unique_ingredients.txt', 'r') as file:
     content = file.readlines()
file.closed

ingredientwords = {}
print('We are starting with %d unique ingredients') % len(content)

#print(content)

for ingredients in content:
    #print item
    words = re.findall(r'\w+', ingredients)
    #print words
    for word in words:
        #ingredientwords[word]=words
        #ingredientwords[word].append(words)
        ingredientwords.setdefault(word,[]).append(words)
        
#print ingredientwords

for key,value in ingredientwords.items():
    ingredientwords[key]=(value, len(value))

numberofingredients={}

for key,value in ingredientwords.items():
    #print("This is how many ingredients contain this word")
    #print key, value[1] 
    numberofingredients[key]=value[1]

print('We know that there are %d unique words in the ingredients') % len(numberofingredients)
    
#unique word against ingredients it appears in
df = pd.DataFrame(ingredientwords.items(), columns=['Ingredient', 'Ingredient words'])
df

# number of times each unique word appears
df2 = pd.DataFrame(numberofingredients.items(), 
                  columns=['Ingredient word', 'Occurences']).sort_values(by=['Occurences'],ascending=False)
df2





```
