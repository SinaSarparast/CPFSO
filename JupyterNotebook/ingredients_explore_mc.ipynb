```python
# work with ingredients to identify most common words /good candidates for stemming and classifying
import re
# Load in the pandas library
import pandas as pd
import numpy as np
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
#from nltk.corpus import wordnet as wn

with open('./Dataset/unique_ingredients.txt', 'r') as file:
     content = file.readlines()
file.closed


ingredientwords = {}
print('We are starting with %d unique ingredients') % len(content)
#print(content)

for ingredients in content:
    #print item
    words = re.findall(r'\w+', ingredients)
    #print words
    for word in words:
        ingredientwords.setdefault(word,[]).append(words)

#dictionary maps every unique word to each ingredient it appears in, with the ingredient itself broken up into a list of words
print ingredientwords

for key,value in ingredientwords.items():
    ingredientwords[key]=(value, len(value))

numberofingredients={}

for key,value in ingredientwords.items():
    #print("This is how many ingredients contain this word")
    #print key, value[1] 
    numberofingredients[key]=value[1]
    
print('We know that there are %d unique words in the ingredients') % len(numberofingredients)

significantwords =[]
insignificantwords=["Oven","it","Cuisine"]
words_appearing_twice_ormore = []

for key,value in numberofingredients.items():
    # assumes words that appear over a certain threshold to be insignificant
    if value >=73:
        insignificantwords.append(str(key))
    # assumes all unique words that appear only once to be significant
    if value<2:
        significantwords.append(str(key))
    elif value >=2:
        words_appearing_twice_ormore.append(str(key))
        
print("I have this many words that appear only once %d" % len(significantwords))
#print significantwords

print("I am processing the words to see whether they are verbs or numbers")

for word in significantwords:
    processed = pos_tag([word])
    print word, processed
    #split out the tag identifying the word type
    tag = ", ".join(t for w, t in processed)
    print(tag)
    if str(tag)=="CD" or str(tag)=="VBN": 
    #or str(tag)=="JJ": including adjectives takes out words like Jamaican but gives us 125 stop words 
        print("I found a number or verb")
        insignificantwords.append(word)
    
print("I have found %d insignificant words" % len(insignificantwords) + "that appear once")
print insignificantwords

for word in words_appearing_twice_ormore:
    processed = pos_tag([word])
    print word, processed
    #split out the tag identifying the word type
    tag = ", ".join(t for w, t in processed)
    print(tag)
    if str(tag)=="CD" or str(tag)=="VBN": 
    #or str(tag)=="JJ": including adjectives takes out words like Jamaican but gives us 125 stop words 
        print("I found a number or verb")
        insignificantwords.append(word)

print("I have found %d insignificant words" % len(insignificantwords))
print insignificantwords

# if these words are removed from the ingredients, will this leave more significant words in the ingredient
# which are better as features?
# what will happen if the remaining ingredient words are stemmed and put back in?
# next steps: do stemming on the words
# map the words back to the dictionary of ingredients, each one must have the final phrase that will be used from it as a feature
# or the item needs to be labelled to say that it is so insignificant it will be completely excluded
# finalphrase:"" ignorecompletely
        
'''
#unique word against ingredients it appears in
#df = pd.DataFrame(ingredientwords.items(), columns=['Ingredient', 'Ingredient words'])
#df

# number of times each unique word appears
df2 = pd.DataFrame(numberofingredients.items(), 
               columns=['Ingredient word', 'Occurences']).sort_values(by=['Occurences'],ascending=False)
df2

'''



```
