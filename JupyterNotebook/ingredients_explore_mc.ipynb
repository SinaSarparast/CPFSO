```python
# work with ingredients to identify most common words /good candidates for stemming and classifying
import re
# Load in the pandas library
import pandas as pd
import numpy as np
import nltk
import pickle
#make sure these library items are downloaded
#nltk.download('punkt')
#nltk.download('averaged_perceptron_tagger')
#nltk.download('wordnet')
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
#from nltk.corpus import wordnet as wn

with open('./Dataset/unique_ingredients.txt', 'r') as file:
     content = file.readlines()
file.closed

ingredientwords = {}
print(('We are starting with %d unique ingredients') % len(content))
#print(content)

with open('./Dataset/unique_cuisines.txt', 'r') as file:
     cuisinescontent = file.readlines()
file.closed

#manually adding country related words in case these are significant
cuisines = ["Belgian"]
for cuisine in cuisinescontent:
    cuisines.append(str(cuisine))

print("here are all the cuisines")
print(cuisines)

ingredienttowords={}

for ingredient in content:
    print(ingredient)
    words = re.findall(r'\w+', ingredient)
    print(words)
    ingredienttowords.setdefault(ingredient,[]).append(words)
    
    #for word in words:
        #ingredient_tuple =  words,word
        #will set dict[key]=default if key is not already in dict.
        #ingredienttowords.setdefault(ingredient,[]).append(ingredient_tuple)
        
        
print("This maps all ingredients to the unique words that make them up")
print(len(ingredienttowords))
print(ingredienttowords)

#write the ingredientwords down to a pkl file
#dictionary maps every unique word to each ingredient it appears in
# to ingredient itself plus the ingredient broken up into a list of words
#print ingredientwords

#do not run this unless needed! writing to workingingredients
inputingredientwords = ingredienttowords
output = open('workingingredients.pkl', 'wb')
pickle.dump(inputingredientwords, output)
output.close()





'''

#unique word against ingredients it appears in
#df = pd.DataFrame(ingredientwords.items(), columns=['Ingredient', 'Ingredient words'])
#df

# number of times each unique word appears
'''
df2 = pd.DataFrame(numberofingredients.items(), 
              columns=['Ingredient word', 'Occurences']).sort_values(by=['Occurences'],ascending=False)
df2
'''
        




```
